### Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры <b>m</b>, <b>n</b>, <b>min_limit</b>, <b>max_limit</b>, где <b>m</b> и <b>n</b> указывают размер
матрицы, а <b>min_lim</b> и <b>max_lim</b> - минимальное и максимальное значение для
генерируемого числа.

```python
import random
import time
```

```python
user_m = int(input())
user_n = int(input())
user_min_limit = int(input())
user_max_limit = int(input())

matrix = []
for _ in range(user_n):
    row = [random.randint(user_min_limit, user_max_limit) for _ in range(user_m)]
    matrix.append(row)
matrix
```

### Задание №3
Реализовать методы сортировки строк числовой матрицы в соответствии с
заданием. Оценить время работы каждого алгоритма сортировки и сравнить его со
временем стандартной функции сортировки. Испытания проводить на сгенерированных
матрицах.

```python
import copy
```

```python
# Сортировка выбором.
# Берётся срез массива, в котором минимальный элемент переносят в самый левый угол,
# после чего срез уменьшается и цикл повторяется.
start_time = time.time()
matrix_copy = matrix

for i in range(len(matrix_copy)):
    for j in range(len(matrix_copy[i])):
        min_index = j
        for k in range(j + 1, len(matrix_copy[i])):
            if matrix_copy[i][k] < matrix_copy[i][min_index]:
                min_index = k
        matrix_copy[i][j], matrix_copy[i][min_index] = matrix_copy[i][min_index], matrix_copy[i][j]
for row in matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка вставкой.
start_time = time.time()
matrix_copy = matrix
for i in range(1, len(matrix_copy)):
    key = matrix_copy[i]
    j = i - 1
    while j >= 0 and matrix_copy[j] > key:
        matrix_copy[j + 1] = matrix_copy[j]
        j -= 1
    matrix_copy[j + 1] = key
for row in matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка обменом. || Сортировка пузырьком.
start_time = time.time()
matrix_copy = matrix
n = len(matrix_copy)
for i in range(n - 1):
    swapped = False
    for j in range(n - 1 - i):
        if matrix_copy[j] > matrix_copy[j + 1]:
            matrix_copy[j], matrix_copy[j + 1] = matrix_copy[j + 1], matrix_copy[j]
            swapped = True
    if not swapped:
        break
for row in matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Сортировка Шелла.
start_time = time.time()
matrix_copy = matrix
n = len(matrix_copy)
gap = n // 2
while gap > 0:
    for i in range(gap, n):
        temp = matrix_copy[i]
        j = i
        while j >= gap and matrix_copy[j - gap] > temp:
            matrix_copy[j] = matrix_copy[j - gap]
            j -= gap
        matrix_copy[j] = temp
    gap //= 2
for row in matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```

```python
# Быстрая сортировка.
start_time = time.time()
matrix_copy = matrix
def quick_sort(matrix_copy):
    if len(matrix_copy) <= 1:
        return matrix_copy
    
    pivot = random.choice(matrix_copy)
    left = [x for x in matrix_copy if x < pivot]
    middle = [x for x in matrix_copy if x == pivot]
    right = [x for x in matrix_copy if x > pivot]
    
    return quick_sort(left) + middle + quick_sort(right)
sort_matrix_copy = quick_sort(matrix_copy)
for row in sort_matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
    
```

```python
# Турнирная сортировка.
start_time = time.time()
matrix_copy = matrix
def build_tournament_tree(leaf_nodes):
    """Создает турнирное дерево для массива"""
    n = len(leaf_nodes)
    tree = [None] * (2 * n)

    for i in range(n):
        tree[n + i] = leaf_nodes[i]

    for i in range(n - 1, 0, -1):
        tree[i] = min(tree[2 * i], tree[2 * i + 1])

    return tree

def update_tournament_tree(tree, index, inf):
    """Обновляет турнирное дерево после удаления минимального элемента"""
    tree[index] = inf

    while index > 1:
        index //= 2
        tree[index] = min(tree[2 * index], tree[2 * index + 1])

def tournament_sort(arr):
    """Турнирная сортировка одномерного списка"""
    if not arr:
        return []
    
    inf = float('inf')
    n = len(arr)
    tree = build_tournament_tree(arr)
    sorted_list = []

    for _ in range(n):
        min_value = tree[1]
        sorted_list.append(min_value)

        index = tree.index(min_value, len(tree) // 2)
        update_tournament_tree(tree, index, inf)

    return sorted_list

def sort_matrix(matrix_copy):
    """Применяет турнирную сортировку к каждой строке матрицы"""
    return [tournament_sort(row) for row in matrix_copy]

sort_matrix_copy = sort_matrix(matrix_copy)
for row in sort_matrix_copy:
    print(row)
print("--- {0} ms ---".format(round((time.time() - start_time)*1000)))
```
